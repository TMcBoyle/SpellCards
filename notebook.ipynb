```python
import drawsvg as draw
import textwrap

def build_centered_text(text, parent_args, style):
    text_x = parent_args['x'] + parent_args['width'] / 2
    text_y = parent_args['y'] + (float(style["font_size"][:-2]) + parent_args['height']) / 2
    return draw.Text(text, x=text_x, y=text_y, **style)

def build_label_text(text, parent_args, style):
    text_x = parent_args['x'] + parent_args['width'] / 2
    text_y = parent_args['y'] - float(style["font_size"][:-2]) / 2
    return draw.Text(text, x=text_x, y=text_y, **style)

# Constants
origin_xy = (-32, -44.5)

width = 241.88977
height = 336.37796

box_wh = {
    "title":    { "width": 226.77167, "height": 30.236221 },
    "wide":     { "width": 211.65355, "height": 15.118111 },
    "large":    { "width": 226.77167, "height": 166.29924 },
    "standard": { "width": 68.031502, "height": 18.897638 },
    "small":    { "width": 18.897638, "height": 18.897638 },
    "tiny":     { "width": 14.000000, "height": 14.000000 },
}

cols = [-24.440945, -16.881889, 54.929138, 134.29922, 158.86617, 183.43311]
rows = [-36.940945,  8.4133873, 31.090555, 49.988194, 80.224411, 106.68111]

reaction_connector_path = f"m -20.661417,27.311026 v 11.338583 h 3.779528"
material_connector_path = f"m 198.55119,27.311026 v 30.236221 h -3.77953"

# Styles
style_background = {
    "fill": "#ffd653"
}
style_box = {
    "fill": "#fafafa",
    "stroke": "#000000",
    "stroke_width": "1.88976",
    "stroke_linejoin": "bevel"
}
style_line = {
    "fill": "none",
    "stroke": "#000000",
    "stroke_width": "1.88976",
    "stroke_linejoin": "bevel"
}
style_header = {
    "font_size": "16pt",
    "font_family": "Cambria",
    "text_align": "justify",
    "text_anchor": "middle",
    "fill": "#000000"
}
style_body = {
    "font_size": "8pt",
    "font_family": "Cambria",
    "text_align": "justify",
    "text_anchor": "middle",
    "fill": "#000000"
}
style_body_big = {
    "font_size": "12pt",
    "font_family": "Cambria",
    "text_align": "justify",
    "text_anchor": "middle",
    "fill": "#000000"
}
style_rules = {
    "font_size": "8pt",
    "font_family": "Cambria",
    "text_align": "justify",
    "text_anchor": "start",
    "fill": "#000000"
}
style_footer = {
    "font_size": "8pt",
    "font_family": "Cambria",
    "text_align": "justify",
    "text_anchor": "end",
    "fill": "#000000"
}

template = draw.Drawing(width, height, origin=origin_xy)

# Background
background = draw.Rectangle(origin_xy[0], origin_xy[1], width, height, **style_background)
template.append(background)

# Spell Name
spell_name_container = draw.Rectangle(cols[0], rows[0], **box_wh["title"], **style_box)
spell_name_value = build_centered_text("FEATHER FALL", spell_name_container.args, style_header)
template.append(spell_name_container)
template.append(spell_name_value)

# Level
level_container = draw.Rectangle(cols[0], rows[0], **box_wh["tiny"], **style_box)
level_value = build_centered_text("1", level_container.args, style_body)
template.append(level_container)
template.append(level_value)

# Casting Time
casting_time_container = draw.Rectangle(cols[0], rows[1], **box_wh["standard"], **style_box)
casting_time_label = build_label_text("Casting Time", casting_time_container.args, style_body)
casting_time_value = build_centered_text("1 reaction", casting_time_container.args, style_body)
template.append(casting_time_label)
template.append(casting_time_container)
template.append(casting_time_value)

# Range
range_container = draw.Rectangle(cols[2], rows[1], **box_wh["standard"], **style_box)
range_label = build_label_text("Range", range_container.args, style_body)
range_value = build_centered_text("60 feet", range_container.args, style_body)
template.append(range_label)
template.append(range_container)
template.append(range_value)

# Components
verbal_container = draw.Rectangle(cols[3], rows[1], **box_wh["small"], **style_box)
somatic_container = draw.Rectangle(cols[4], rows[1],  **box_wh["small"], **style_box)
material_container = draw.Rectangle(cols[5], rows[1], **box_wh["small"], **style_box)
component_label = build_label_text("Components", somatic_container.args, style_body)
verbal_value = build_centered_text("V", verbal_container.args, style_body_big)
somatic_value = build_centered_text("S", somatic_container.args, style_body_big)
material_value = build_centered_text("M", material_container.args, style_body_big)
template.append(component_label)
template.append(verbal_container)
template.append(verbal_value)
template.append(somatic_container)
template.append(somatic_value)
template.append(material_container)
template.append(material_value)

# Reaction Condition
reaction_condition_container = draw.Rectangle(cols[1], rows[2], **box_wh["wide"], **style_box)
reaction_condition_connector = draw.Path(reaction_connector_path, **style_line)
reaction_condition_value = build_centered_text(
    "When  you or a creature within 60 feet falls.",
    reaction_condition_container.args,
    style_body
)
template.append(reaction_condition_container)
template.append(reaction_condition_connector)
template.append(reaction_condition_value)

# Specific Material
specific_material_container = draw.Rectangle(cols[1], rows[3], **box_wh["wide"], **style_box)
specific_material_connector = draw.Path(material_connector_path, **style_line)
specific_material_value = build_centered_text(
    "A small feather or piece of down.",
    specific_material_container.args,
    style_body
)
template.append(specific_material_container)
template.append(specific_material_connector)
template.append(specific_material_value)

# Duration
duration_container = draw.Rectangle(cols[0], rows[4], **box_wh["standard"], **style_box)
duration_label = build_label_text("Duration", duration_container.args, style_body)
duration_value = build_centered_text("1 minute", duration_container.args, style_body)
template.append(duration_label)
template.append(duration_container)
template.append(duration_value)

# Attributes
ritual_container = draw.Rectangle(cols[2], rows[4], **box_wh["standard"], **style_box)
concentration_container = draw.Rectangle(cols[3], rows[4], **box_wh["standard"], **style_box)
attributes_label = build_label_text(
    "Attributes", 
    {
        'x': (cols[2] + cols[3] + box_wh["standard"]["width"]) / 2,
        'y': rows[4],
        'width': 0,
        'height': 0
    },
    style_body
)
ritual_value = build_centered_text("Ritual", ritual_container.args, style_body)
concentration_value = build_centered_text("Concentration", concentration_container.args, style_body)
template.append(attributes_label)
template.append(ritual_container)
template.append(ritual_value)
template.append(concentration_container)
template.append(concentration_value)

# Spell Details
spell_details_container = draw.Rectangle(cols[0], rows[5], **box_wh["large"], **style_box)
spell_details_text = "Choose up to five falling creatures within range. A falling creature's rate of " \
    "descent slows to 60 feet per round until the spell ends. If the creature lands " \
    "before the spell ends, it takes no falling damage and can land on its feet, and " \
    "the spell ends for that creature."
spell_details_text = textwrap.wrap(spell_details_text, 50)
spell_details_value = draw.Text(spell_details_text, x=-16.882812, y=123.87047, **style_rules)
template.append(spell_details_container)
template.append(spell_details_value)

# Footer
footer = draw.Text("CONJURATION", x=203, y=285.35736, **style_footer)
template.append(footer)

template
```

```python
import card_builder
import card_loader

def generate_cards(caster: str, levels: list[str], colour="#ffffff", save=False):
    data = card_loader.load_card_data("Revised Spells.csv")
    data = data[data["casters"].str.contains(caster, na=False)]
    data = data[data["level"].isin(levels)]

    spell_dicts = data.drop(["casters"], axis=1).to_dict(orient="records")
    svgs = []
    for idx, s in enumerate(spell_dicts):
        spell = card_builder.Spell(**s)
        card = card_builder.Card(spell)
        svgs.append(card.to_svg(colour))

        if save:
            svgs[-1].save_svg(f"{caster}/" + f"{spell.name}.svg".replace("/", "_"))
    return svgs

svgs = []
#svgs.extend(generate_cards("Cleric", ["C", "1", "2"], "#e6c03b"))
#svgs.extend(generate_cards("Druid", ["C", "1", "2"], "#b39eb5"))
svgs.extend(generate_cards("Sorcerer", ["C", "1", "2"], "#ff964f"))
```

```python
import drawsvg as draw
from card_builder import Card
import textwrap

def build_pages(page_width, page_height, cards, spacing=12):
    card_width = Card.width + spacing
    card_height = Card.height + spacing
    
    cols = int(page_width // card_width)
    rows = int(page_height // card_height)
    
    cards_per_page = cols * rows

    idx = 0
    pages = []
    for _ in range(int(len(cards) // cards_per_page) + 1):
        pages.append(draw.Drawing(page_width, page_height))
        page_idx = idx % cards_per_page
        while page_idx < cards_per_page and idx < len(cards):
            for element in cards[idx].elements:
                pages[-1].append(draw.Use(element, spacing + card_width * (page_idx % cols), spacing + card_height * (page_idx // rows)))
            idx += 1
            page_idx += 1

    return pages

pages: list[draw.Drawing] = build_pages(Card.width * 3 + 50, Card.height * 3 + 50, svgs)

for n, page in enumerate(pages):
    page.save_svg(f"cards/page{n}.svg")
```
